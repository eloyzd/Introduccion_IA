```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.model_selection import train_test_split
```

```python
def build_dataset(n,k):
  x=np.arange(0,n) + 0.8
  y = np.random.rand(n)*k
  return x, y

def Regression_pack(n,k,train_size) :
  m = 2
  h = 0

  x,y = build_dataset(n,k )

  x = np.array(x)
  y = np.array(y)
  x = x.reshape(len(x),1)
  x = x.reshape(len(y),1)


  x_train, x_test, y_train, y_test = train_test_split(x,y,train_size = TRAIN_SIZE)

  alpha = 0.5
  L = LinearRegression()
  L.fit(x_train,y_train)
  L1 = Lasso(alpha = alpha)
  L1.fit(x_train,y_train)
  L2 = Ridge(alpha = alpha)
  L2.fit(x_train,y_train)

  return x_train, x_test, y_train, y_test, L, L1, L2
```

```python
n=100
TRAIN_SIZE = 0.8
for i in range(3):
  x_train, x_test, y_train, y_test, L, L1, L2 = Regression_pack(n,1,TRAIN_SIZE)
  plt.figure(figsize=(10,3))
  plt.subplot(121)
  plt.plot(x_train,y_train,'o')
  plt.plot(x_train,L.predict(x_train),'c')
  plt.plot(x_train,L1.predict(x_train),'m')
  plt.plot(x_train,L2.predict(x_train),'g')
  plt.subplot(122)
  plt.plot(x_test,y_test,'ok',alpha=0.5)
  plt.plot(x_test,L.predict(x_test),'c')
  plt.plot(x_test,L1.predict(x_test),'m')
  plt.plot(x_test,L2.predict(x_test),'g')
  plt.show()
  
```
