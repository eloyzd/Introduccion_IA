# Ejecicio 1: Normalización
Muchos algoritmos de Machine Learning necesitan datos de entrada centrados y normalizados. Una normalización habitual es el z-score, que implica restarle la media y dividir por el desvío a cada feature de mi dataset.

Dado un dataset X de n muestras y m features, implementar un método en numpy para normalizar con z-score. Pueden utilizar np.mean() y np.std().

```python
import numpy as np
n = 20 
m = 5
X = np.random.rand(n,m)

def z_score(X):
  mu = np.mean(X,axis=0)
  var = np.std(X,axis=0)
  X_ = X - mu
  X_ = X_/var
  return X_

z_score(X)
```

# Ejecicio 2: Remover filas y columnas con NaNs en un dataset
Dado un dataset, hacer una función que, utilizando numpy, filtre las columnas y las filas que tienen NaNs.


```python
import numpy as np

def delete_nan(data):
  col = np.where(np.isnan(np.sum(data,axis=0)),1,0)
  row = np.where(np.isnan(np.sum(data,axis=1)),1,0)
  idx = range(len(col))
  filt_col = [i for i in idx if col[i] == 0]
  idx = range(len(row))
  filt_raw = [i for i in idx if row[i] == 0]
  data_ =data[:,filt_col]
  return data_[filt_raw]

n = 20 
m = 5
np.random.seed(0)
data = np.random.randn(n, m) 

index_nan = np.random.choice(data.size, 2, replace=False) 
data.ravel()[index_nan] = np.nan 
print(data)

delete_nan(data)
```

```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.linear_model import LinearRegression

class Data:
  def __init__(self):
    pass

  def load(self,path, delimiter = ';', header = None):
    self.data = pd.read_csv(path, delimiter = delimiter, header = header)

class Metric(object):
    def __call__(self, target, prediction):
        return NotImplemented


class MSE(Metric):
    def __init__(self):
        Metric.__init__(self)

    def __call__(self, target, prediction):
        n = target.size
        return np.sum((target - prediction) ** 2) / n

class Accuracy(Metric):
    def __call__(self, target, prediction):
        return np.mean(target == prediction, axis=0)


class Precision(Metric):
    def __call__(self, target, prediction):
        inverse_truth = np.where(target == 0, 1, 0)
        inverse_predictions = np.where(prediction == 0, 1, 0)
        TP = np.sum(prediction * target)
        TN = np.sum(inverse_predictions * inverse_truth)
        FN = np.sum(target * inverse_predictions)
        FP = np.sum(prediction * inverse_truth)
        return TP / (TP + FP)


class Recall(Metric):
    def __call__(self, target, prediction):
        inverse_truth = np.where(target == 0, 1, 0)
        inverse_predictions = np.where(prediction == 0, 1, 0)
        TP = np.sum(prediction * target)
        TN = np.sum(inverse_predictions * inverse_truth)
        FN = np.sum(target * inverse_predictions)
        FP = np.sum(prediction * inverse_truth)
        return TP / (TP + FN)


def split_data(X,Y ):
  idx = np.random.permutation(len(X))
  x_train = X[idx[0:int(len(idx)*0.8)]]
  x_test = X[idx[int(len(idx)*0.8)::]]
  y_train = Y[idx[0:int(len(idx)*0.8)]]
  y_test = Y[idx[int(len(idx)*0.8)::]]
  return x_train, x_test, y_train, y_test


def delete_nan(data):
  data = np.array(data)
  col = np.where(np.isnan(np.sum(data,axis=0)),1,0)
  row = np.where(np.isnan(np.sum(data,axis=1)),1,0)
  idx = range(len(col))
  filt_col = [i for i in idx if col[i] == 0]
  idx = range(len(row))
  filt_raw = [i for i in idx if row[i] == 0]
  data_ =data[:,filt_col]
  return data_[filt_raw]

def mean_nan(data):
  data = np.array(data)
  row = np.where(np.isnan(np.sum(data,axis=1)),1,0)
  filt_raw = [i for i in range(len(row)) if row[i] == 0]
  data_no_nan = data[filt_raw]
  means = np.mean(data_no_nan,axis=0)
  mask = np.isnan(data)
  mean_mtx = np.ones_like(data)
  mean_mtx[:]=means
  for i in range(len(data)):
    for j in range(len(data[0])):
      if np.isnan(data[i][j]):
        data[i][j] = mean_mtx[i][j]
      else:
        pass
  return data

data = Data()
data.load(path = '/content/clase3v2.csv')
```

```python
data_fixed_1 = delete_nan(data.data)
data_fixed_2 = mean_nan(data.data)

x_1 = np.array(data_fixed_1)[:,0:-1]
y_1 = np.array(data_fixed_1)[:,-1::]

x_2 = np.array(data_fixed_2)[:,0:-1]
y_2 = np.array(data_fixed_2)[:,-1::]

x_train_1, x_test_1, y_train_1, y_test_1 = split_data(x_1, y_1)
x_train_2, x_test_2, y_train_2, y_test_2 = split_data(x_2, y_2)


```

```python
pca_1 = PCA(n_components = 3).fit(x_train_1)
pca_2 = PCA(n_components = 3).fit(x_train_2)
```

```python
x_train_1_pca = pca_1.transform(x_train_1)
x_train_2_pca = pca_2.transform(x_train_2)

x_test_1_pca = pca_1.transform(x_test_1)
x_test_2_pca = pca_2.transform(x_test_2)
```

```python

LR_1 = LinearRegression()
LR_1.fit(x_train_1_pca,y_train_1)

LR_2 = LinearRegression()
LR_2.fit(x_train_2_pca,y_train_2)
```

```python

y_test_1_pred = LR_1.predict(x_test_1_pca)
y_test_2_pred = LR_2.predict(x_test_2_pca)

```

```python
mse_1 = MSE()
mse_2 = MSE()
print(mse_1(y_test_1,y_test_1_pred))
print(mse_2(y_test_2,y_test_2_pred))
```
